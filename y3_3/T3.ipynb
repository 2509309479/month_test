```python
# **1.	读取数据新闻.xlsx(3分)
data = pd.read_excel("新闻.xlsx")
titles = data["标题"]
data
```

```python
# **2.	遍历来进行所有标题的分词(3分)
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_df=1.0, min_df=1, max_features=None)
X = vectorizer.fit_transform(titles)
for name in vectorizer.get_feature_names_out():
    print(name)
```
